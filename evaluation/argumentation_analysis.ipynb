## Display of Results
In the following the method to display the results will be shown.

```python
import matplotlib.pyplot as plt
import numpy as np
from ArgumentationAnalysisResult import COMPONENT_NAMES

# Set global font sizes
plt.rcParams.update({
    'axes.titlesize': 18,
    'axes.labelsize': 16,
    'xtick.labelsize': 18,
    'ytick.labelsize': 16,
    'legend.fontsize': 14,
    'legend.title_fontsize': 16
})

# Components in order

labels = ["PRESENT", "IMPLICT_WEAK", "ABSENT"]

# Prepare aggregated data
model_component_stats = {}
component_names = [str(component).lower() for component in COMPONENT_NAMES]

classification_score = {
    "PRESENT": 3,
    "IMPLICT_WEAK": 2,
    "ABSENT": 1
}

def plot_argumentation_anaylisis_results(all_models_data):
    average_scores = {component: [] for component in component_names}
    model_names = list(all_models_data.keys())

    for component in component_names:
        for model in model_names:
            scores = []
            for result in all_models_data[model]:
                status = getattr(result, component)
                scores.append(classification_score[status.name])
            average = sum(scores) / len(scores) if scores else 0
            average_scores[component].append(average)
        model_names = list(all_models_data.keys())

    n_models = len(model_names)
    n_components = len(COMPONENT_NAMES)
    bar_width = 0.12
    index = np.arange(n_models)

    # Reuse average_scores from earlier
    fig, ax = plt.subplots(figsize=(10, 6))
    colors = plt.cm.tab10(np.linspace(0, 1, n_components))

    # Adjust bar positions and draw bars for each Toulmin component
    for i, component in enumerate(component_names):
        bar_positions = index + (i - n_components / 2) * bar_width + bar_width / 2
        ax.bar(bar_positions, average_scores[component], width=bar_width, label=f"{component.capitalize()}", color=colors[i])

    ax.set_xlabel("")
    ax.set_ylabel("Average Dimension Score")
    ax.set_title("Per-Dimension Averages per Model (Toulmin Analysis)")
    ax.set_xticks(index)
    ax.set_xticklabels(model_names)
    ax.legend(title="Component", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)  # or rotation=30 for shallower angle, 60 for steeper
    plt.tight_layout()
    plt.show()
```

### Results of 1. Idea (Good/Simple)

```python
all_models_data = {
    "Mistral": MISTRAL_ANALYSES,
    "Mixtral": MIXTRAL_ANALYSIS_RESULTS,
    "Deepseek-R1": DEEPSEEK_ANALYSIS_RESULTS,
    "Gemma3": GEMMA3_ANALYSIS_RESULTS,
    "Llama3.1": LLAMA3_1_ANALYSIS_RESULTS,
    "Phi": PHI_ANALYSIS_RESULTS,
}

plot_argumentation_anaylisis_results(all_models_data)
```
